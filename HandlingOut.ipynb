### Importing the file

```python
import numpy as np

# Specify the path to your text file
file_path = r"C:\Users\omri.c\OneDrive - arberobotics\Desktop\Omri\University\מעבדה מערכות אוטונומיות\Files From The Lab\out.txt"
```

### Getting the position and angle of each iterate

```python
# Read the text file
with open(file_path, 'r') as file:
    lines = file.readlines()

# Initialize empty lists to store the updated positions and angles
positions = []
angles = []

# Extract positions and angles from the lines
for line in lines:
    if 'Update pos_x and pos_y:' in line:
        # Extract the pos_x and pos_y values
        values = line.split(':')[1].strip().split('|')[0].strip()
        x, y = map(float, values[1:-1].split(','))
        positions.append([x, y])
        
        # Extract the angle value
        angle_value = line.split('-')[-1].strip()
        angles.append(float(angle_value))

# Convert the lists of positions and angles to NumPy arrays
positions_array = np.array(positions)
angles_array = np.array(angles)

# Print the resulting positions and angles arrays
print("Positions:")
print(positions_array)
print("Angles:")
print(angles_array)
```

```python
import re

# Initialize empty lists to store the extracted values
point_numbers = []
remaining_dists = []
target_velocities_left = []
target_velocities_right = []
dist_to_next_point = []

# Extract values from the lines
for line in lines:
    if 'Reached point number' in line:
        point_number = int(line.split()[-1])
        point_numbers.append(point_number)
    
    if 'Remaining dist =' in line:
        remaining_dist = float(line.split()[-1])
        remaining_dists.append(remaining_dist)
    
    if 'target_velocity_left' in line:
        match = re.search(r'target_velocity_left\s*-\s*([\d.]+)', line)
        if match:
            target_velocity_left = float(match.group(1))
            target_velocities_left.append(target_velocity_left)

    if 'target_velocity_right' in line:
        match = re.search(r'target_velocity_right\s*-\s*([\d.]+)', line)
        if match:
            target_velocity_right = float(match.group(1))
            target_velocities_right.append(target_velocity_right)
    
    if 'dist -' in line:
        match = re.search(r"dist - ([\d.]+)", line)
        if match:
            dist = float(match.group(1))
            dist_to_next_point.append(dist)


# Convert the lists to NumPy arrays
point_numbers_array = np.array(point_numbers)
remaining_dists_array = np.array(remaining_dists)
target_velocities_left_array = np.array(target_velocities_left)
target_velocities_right_array = np.array(target_velocities_right)
dist_to_next_point_array = np.array(dist_to_next_point)

# Print the resulting arrays
print("Point Numbers:")
print(point_numbers_array)
print("Remaining Distances:")
print(remaining_dists_array)
print("Target Velocities (Left):")
print(target_velocities_left_array)
print("Target Velocities (Right):")
print(target_velocities_right_array)
print("dist to next point ")
print(dist_to_next_point_array)
```

### Showing desired angle and Velocity diff
